```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iot_sensor_data.csv")
```

```python
print("Columns in dataset:")
print(df.columns)

print("\nFirst 5 rows:")
df.head()
```

```python
timestamp = df["Timestamp"]
temperature = df["Temperature(C)"]
humidity = df["Humidity(%)"]
light_intensity = df["Light_Intensity(lx)"]
motion = df["Motion"]
```

```python
print("Temperature values:")
print(temperature)

print("\nHumidity values:")
print(humidity)

print("\nLight Intensity values:")
print(light_intensity)

print("\nMotion values:")
print(motion)
```

```python
df.describe()
```

```python
df.isnull().sum()
df = df.fillna(df.mean(numeric_only=True))
```

```python
plt.plot(temperature)
plt.title("Temperature Sensor Readings")
plt.xlabel("Samples")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
plt.plot(humidity)
plt.title("Humidity Sensor Readings")
plt.xlabel("Samples")
plt.ylabel("Humidity (%)")
plt.show()
```

```python
df.to_csv("iot_sensor_cleaned.csv", index=False)
print("Cleaned IoT sensor data saved successfully")
```
